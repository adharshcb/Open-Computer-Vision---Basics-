
# Morphological Transformations

### Morphological transformations are some simple operations based on the image shape.It is normally performed on binary images.This needs a structuring element or a kernal that tells we how to change the value of any given pixel by combinig it with different amounts of the neighboring pixels.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('smarties.png',0)
# read image in binary

_,mask = cv2.threshold(img,220, 255,cv2.THRESH_BINARY_INV)
# threshold mask

kernal = np.ones((2,2),np.uint8) 

dilation = cv2.dilate(mask,kernal,iterations=2) 
# args               (source,kernal,number of iterations)
# the pixel element will be 1 if atleast 1 pixel under the kernal is one.

erosion = cv2.erode(mask,kernal,iterations=1) # 
#args              (source,kernal,number of interations)
# the pixel element will be 1 only if all the pixel element under the kernal is 1.

opening = cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernal)
# args                    (source,operaion,kernal)
# erosion followed by dilation

closing = cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernal)
# dilation followed by erosion

mg = cv2.morphologyEx(mask,cv2.MORPH_GRADIENT,kernal)
# morphological gradient - difference between dilation and erosion 

th = cv2.morphologyEx(mask,cv2.MORPH_TOPHAT,kernal)
# tophat - difference between image and opening 

title = ['image','mask','dilation','erosion','opening','closing','mg','th']
images = [img,mask,dilation,erosion,opening,closing,mg,th]

for i in range(8):
    plt.subplot(2,4,i+1),plt.imshow(images[i],'gray')
    plt.title(title[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()
```
