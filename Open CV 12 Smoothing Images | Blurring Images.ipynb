## Homogeneous filter

### In this filter each output pixel is the mean of its kernal neighbors.

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread('Noise_salt_and_pepper.png')
# img = cv2.medianBlur(img,5)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

kernal = np.ones((5,5),np.float32)/25
dest = cv2.filter2D(img,-1,kernal)

blur = cv2.blur(img,(5,5))
```

### In case of one-dimensional signals,images can be filtered using low-pass filters(LPF),high-pass filters(HPF) etc.

## LPF helps in removing noises, blurring the images.
## HPF helps in finding edges in the images.

# Gaussian filter

### Gaussian filter is using different weight kernal in both x and y direction.
for removing high frequency noise from images

```python
gblur = cv2.GaussianBlur(img,(5,5),0)
```

# Median Filter

Median filter is something that replace each pixel's value with the median of its neighboring pixels.Good for salt and pepper noise.

```python
img2 = cv2.imread('Noise_salt_and_pepper.png')
median = cv2.medianBlur(img2,5)
#                       (source,kernal-size(must be odd except 1))
```

# Bilateral filter

Bilateral filter is highly effective in noise removal while keeping the edge sharp.

```python
bilateralFilter = cv2.bilateralFilter(img2,9,75,75)
```

```python
titles = ['median','2D Convolution','blur','gaussianBlur','image2','median','bilateral']
images = [img,dest,blur,gblur,img2,median,bilateralFilter]

for i in range(len(images)):
    plt.subplot(2,4,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```
