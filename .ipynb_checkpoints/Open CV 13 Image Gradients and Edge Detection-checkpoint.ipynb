```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Image Gradients

An image gradient is a directional change in the intensity or color in an image.

## Laplacian Gradient

```python
img = cv2.imread('sudoku.png',0)
lap = cv2.Laplacian(img,cv2.CV_64F,ksize = 3)
#                  (source,datatype(64f supports negative number))
lap = np.uint8(np.absolute(lap))
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0)
# changes in vertical direction

sobely = cv2.Sobel(img,cv2.CV_64F,0,1)
# directional change in intensity of horizontal direction

sobelx = np.uint8(np.absolute(sobelx))
sobely = np.uint8(np.absolute(sobely))

sobelCombine = cv2.bitwise_or(sobelx,sobely)# combine sobelx and sobely
```

```python
titles = ['image','lap','sobelx','sobely','sobel combine']
images = [img,lap,sobelx,sobely,sobelCombine]

for i in range(len(images)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```
